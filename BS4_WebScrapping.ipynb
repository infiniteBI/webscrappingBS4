```python
#import libraries
from bs4 import BeautifulSoup
import requests
import csv
import datetime
```

```python
URL= 'https://www.scrapethissite.com/pages/simple'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/133.0.0.0 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
page_scraping = requests.get(URL, headers= headers)
soup= BeautifulSoup(page_scraping.content, "html.parser")
country_names = soup.findAll('h3', attrs={'class':"country-name"})
country_capitals = soup.findAll('span', attrs={'class':"country-capital"})
country_populations = soup.findAll('span', attrs={'class':"country-population"})
country_areas = soup.findAll('span', attrs={'class':"country-area"})
```

```python
#Let's remove the nested class i in the country name, capital, population and area

country = []
capital = []
population= []
area= []
for item in country_names:
    country_name = item.text.strip()
    country.append(country_name)
for item in country_capitals:
    country_capital = item.text.strip()
    capital.append(country_capital)
for item in country_populations:
    country_population = item.text.strip()
    population.append(country_population)
for item in country_areas:
    country_area = item.text.strip()
    area.append(country_area)
```

```python
 #let's store the data in a csv file

first_line= ['Country', 'Capital', 'Population','Area']
data_row= [country, capital, population, area]

file = open ('countryscrapping.csv', 'w', newline='', encoding='UTF8')
writer = csv.writer(file)
writer.writerow(first_line)

for count, cap, pop, ar in zip(country, capital, population, area):
    print(count + "-" + cap + "-" + pop + "-" + ar)
    writer.writerow([count, cap, pop, ar])

file.close()
```

```python
import pandas as pd

df = pd.read_csv(r'countryscrapping.csv')
```

```python
density_values = [] 

for pop, ar in zip(df['Population'], df['Area']):
    try:
        den = float(pop) / float(ar)
        density_values.append(round(den, 3))
    except ZeroDivisionError:
        density_values.append('') 

df['Density'] = density_values
```

```python
updated_file = 'countryscrapping.csv'
df.to_csv(updated_file, index= False)
```
